```python
import pandas as pd
import numpy as np
import os.path as op
from matplotlib import pyplot as plt

IMAGE_WIDTH = {5: 15, 20: 60, 60: 180}
IMAGE_HEIGHT = {5: 32, 20: 64, 60: 96}


```

```python
for year in range(1993, 2020):
    # Load images by year
    images = np.memmap(op.join("./img_data", "./monthly_20d", f"20d_month_has_vb_[20]_ma_{year}_images.dat"), dtype=np.uint8, mode='r').reshape(
                            (-1, IMAGE_HEIGHT[20], IMAGE_WIDTH[20]))
    print(f"Images shape for year {year}: ", images.shape)
    print("First image for year", year)
    print(images[0])
    print()
    
    # Load Labels
    label_df = pd.read_feather(op.join("./img_data", "./monthly_20d", f"20d_month_has_vb_[20]_ma_{year}_labels_w_delay.feather"))
    print("Label DataFrame for year", year)
    print(label_df.head())
    print()
    
    # Plot the sample images
    for i in range(5):
        print("Sample image", i, "for year", year)
        plt.imshow(images[i], cmap='gray')
        plt.show()
```
